```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()

```

```python
df.info()
df.shape
df.columns
```

```python
df.isnull().sum()
```

```python
df['country'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Available', inplace=True)
df['date_added'].fillna('Not Available', inplace=True)
df['director'].fillna('Unknown', inplace=True)
df['rating'].fillna('Not Available', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='listed_in', data=df, order=df['listed_in'].value_counts().index[:10])
plt.title("Top 10 Genres on Netflix")
plt.show()
```

```python
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar', figsize=(8,5), color='orange')
plt.title("Top 10 Content-Producing Countries on Netflix")
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10,5))
plt.title("Number of Releases per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```
