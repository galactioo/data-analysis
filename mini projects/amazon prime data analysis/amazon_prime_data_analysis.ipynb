```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("amazon_prime_titles.csv")
df.head()
```

```python
df.info()
df.shape
df.columns
```

```python
df.isnull().sum()
```

```python
df['director'].fillna("Not Available", inplace=True)
df['cast'].fillna("Not Available", inplace=True)
df['country'].fillna("Unknown", inplace=True)
df['rating'].fillna("Not Rated", inplace=True)

```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type')
plt.title("Movies vs TV Shows on Amazon Prime")
plt.show()

```

```python
plt.figure(figsize=(10,6))
sns.countplot(
    y=df["listed_in"],
    order=df["listed_in"].value_counts().head(10).index
)
plt.title("Top 10 Genres on Amazon Prime")
plt.show()
```

```python
df["release_year"].value_counts().sort_index().plot(figsize=(10,6),kind="line")
plt.title("Amazon Prime Content Added Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python

df["country"].value_counts().head(10).plot(
    kind="bar",
    figsize=(12,6)
)
plt.title("Top 10 Content-Producing Countries on Amazon Prime")
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    data=df,
    y='rating',
    order=df['rating'].value_counts().index
)
plt.title("Distribution of Ratings")
plt.show()
```
